# **List**

A list in Python is a collection data type that is ordered, **mutable** (changeable), and **allows duplicate elements**. Lists are used to store multiple items in a single variable. Here are some key characteristics and functionalities of lists in Python:

## **Key Characteristics of List:**
- Ordered: Lists maintain the order of elements, meaning that the position of each element is fixed.
- Mutable: you can modify list elements.
- Heterogeneous: Lists can store elements of different data types (e.g., integers, strings, lists, etc.).
- Indexable: Elements in a list can be accessed using indices, starting from 0 for the first element.

## **Creating a List**

You can create a list by placing comma-separated values inside square brackets `[]`.


````python
# Example of creating a list
fruits = ["apple", "banana", "cherry"]
print(fruits)
````

## Empty list

````python
empty_list = []
print(empty_list)
````

## **Accessing List Items**

List items are indexed, with the first item having an index of 0. You can access list items using the index.

````python
# Accessing the first item
fruits = ["apple", "banana", "cherry"]
print(fruits[0])  # Output: apple

# Accessing the last item using negative indexing
print(fruits[-1])  # Output: cherry
````

*Note:*
1. It is the same as accessing each character in a string using an index number.
1. Slicing also applied to a list.

````python
fruits = ["apple", "banana", "cherry"]
print(fruits[0:2])  # Output: ['apple', 'banana']
````

## Index out of range

````python
fruits = ["apple", "banana", "cherry"]
print(fruits[3])  # Output: IndexError: list index out of range
````

## Built-in Python's functions that always applied to a list



````python
numbers = [1, 2, 3, 4, 5]
print(len(numbers))
print(sum(numbers))
print(max(numbers))
print(min(numbers))
````

## List of List

````python
matrix = [[1, 2], [3, 4]]
print(matrix)  # Output: [[1, 2], [3, 4]]
print('_' * 20)

#access each element
for row in matrix:
    for item in row:
        print(item)
print('_' * 20)

#access each element using index
for i in range(len(matrix)):
    for j in range(len(matrix[i])):
      print(matrix[i][j], end = " ")
    print()

print('_' * 20)
for col1, col2 in matrix:
  print(col1, col2)
````

##Assigning a list

When you assign a list to a new variable, you're not creating a copy. Both variables reference the same list in memory. If you change one, the other is affected as well.

````python
list1 = [1,2,3,4,5]
list2 = list1
list2[0] = 6
print(list1)
````

From the above code, the first index in `list2` was changed, but `list1` and `list2` point to the same list, so the change also affects `list1`.

## **List functions**

Python lists come with a variety of built-in methods for manipulating list items. Here are some of them.

1. `append()`

```
def append(self, element) -> None:
    """
    Append an element to the end of the list.

    Parameters:
    element: The item to be added to the list.

    Returns:
    None
    """
```

- self: Refers to the list object.
- element: The item to be appended to the end of the list.
- Returns: This method does not return any value (None).

````python
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")
print(fruits)  # Output: ['apple', 'banana', 'cherry', 'orange']
````

2. `insert()`

```
def insert(self, index: int, element) -> None:
    """
    Insert an element at the specified position.

    Parameters:
    index (int): The position at which the element should be inserted.
    element: The item to be inserted into the list.

    Returns:
    None
    """
```

- self: Refers to the list object.
- index: The position at which to insert the new element.
- element: The item to be inserted.
- Returns: This method does not return any value (None).


````python
fruits = ["apple", "banana", "cherry"]
fruits.insert(1, "kiwi")
print(fruits)  # Output: ['apple', 'kiwi', 'banana', 'cherry']
````

3. `remove()`

```
def remove(self, element) -> None:

```

4. `pop()`

```
def pop(self, index: int = -1) -> Any:
    """
    Remove and return the element at the specified position.
    If no index is specified, remove and return the last item in the list.

    Parameters:
    index (int, optional): The position of the item to be removed. Default is -1.

    Returns:
    Any: The item that was removed from the list.
    """

```
- self: Refers to the list object.
- index: The position of the item to be removed. If not specified, the last item is removed.
- Returns: The removed item.

````python
fruits = ["apple", "banana", "cherry"]
last_fruit = fruits.pop()
print(last_fruit)  # Output: cherry
print(fruits)  # Output: ['apple', 'banana']
````

5. `sort()`

```
def sort(self, *, key: Callable[[Any], Any] = None, reverse: bool = False) -> None:
    """
    Sort the list in ascending order. Use the `key` function for custom sorting.
    Set `reverse` to True to sort in descending order.

    Parameters:
    key (Callable[[Any], Any], optional):
    
    A function of one argument that is used to extract a comparison key from each list element. Default is None.
    
    reverse (bool, optional): If True, the list elements are sorted
    as if each comparison were reversed. Default is False.

    Returns:
    None
    """

```
- self: Refers to the list object.

- *: This asterisk signifies that the following parameters (key and reverse) must be specified using their keyword names.

- key: A function of one argument that is used to extract a comparison key from each list element. This allows for custom sorting. The default is None.

- reverse: A boolean flag. If set to True, the list is sorted in descending order. The default is False.


````python
fruits = ["apple", "cherry", "banana"]
fruits.sort()
print(fruits)  # Output: ['apple', 'banana', 'cherry']
````

````python
fruits = ["apple", "cherry", "banana"]
fruits.sort(key=len)  #sort by the length of the string, key=len means applies key() function to each string
print(fruits)  # Output: ['apple', 'cherry', 'banana'] => 'cherry' and 'banana' have the same length, so use their original order.
````

````python
fruits = ["apple", "cherry", "banana"]
fruits.sort(key=len, reverse=True)  #sort by the length of the string in descending order
print(fruits)  # Output: ['cherry', 'banana', 'apple']
````

7. `extend()`

```

```

As discussed in the previous chapter, the string is also an iterable object (you can iterate over each character in a string). That's why each character in the string `str` is added to the `fruits` list.

8. `index()`

The index() function in Python is a method used to find the index of the first occurrence of a specified element in a list. If the element is not found, it raises a ValueError.

```
def index(self, element: Any, start: int = 0, end: int = None) -> int:
    """
    Return the first index of a value in the list.

    Parameters:
    element (Any): The item to search for in the list.
    start (int, optional): The starting index from where to begin the search. Default is 0.
    end (int, optional): The ending index up to which to search. Default is None, meaning the end of the list.

    Returns:
    int: The index of the first occurrence of the element.

    Raises:
    ValueError: If the element is not found in the list.
    """

```
**Parameters:**
- self: Refers to the list object.
- element: The item whose index is to be found.
- start: (Optional) The position from where the search should begin. Default is 0.
- end: (Optional) The position where the search should end. Default is None, which means the search goes until the end of the list.

**Return Value:**
- Returns the index of the first occurrence of the specified element in the list.
**Exceptions:**
- Raises a ValueError if the element is not found in the list.

*Note:*

Exception will be discussed later

````python
# Example list
my_list = [10, 20, 30, 20, 40, 50]

# Find the index of the first occurrence of 20
index_20 = my_list.index(20)
print("Index of 20:", index_20)  # Output: Index of 20: 1

# Find the index of 20, starting from index 2
index_20_from_2 = my_list.index(20, 2)
print("Index of 20 starting from index 2:", index_20_from_2)  # Output: Index of 20 starting from index 2: 3

# Attempting to find an element not in the list
try:
    index_100 = my_list.index(100)
except ValueError as e:
    print("Error:", e)  # Output: Error: 100 is not in list
````

9. `count()`

The count() function in Python is a method used to count the number of occurrences of a specified element in a list. It returns an integer representing the number of times the specified element appears in the list.

```
def count(self, element: Any) -> int:
    """
    Return the number of occurrences of a value in the list.

    Parameters:
    element (Any): The item whose occurrences in the list need to be counted.

    Returns:
    int: The number of times the specified element appears in the list.
    """

```
**Parameters:**
- self: Refers to the list object.
- element: The item whose occurrences are to be counted.

**Return Value:**
- Returns an integer representing the number of occurrences of the specified element in the list.

````python
# Example list
my_list = [1, 2, 3, 2, 4, 2, 5]

# Count the number of times 2 appears in the list
count_2 = my_list.count(2)
print("Number of times 2 appears:", count_2)  # Output: Number of times 2 appears: 3

# Count the number of times 5 appears in the list
count_5 = my_list.count(5)
print("Number of times 5 appears:", count_5)  # Output: Number of times 5 appears: 1

# Count the number of times 10 appears in the list (which is not present)
count_10 = my_list.count(10)
print("Number of times 10 appears:", count_10)  # Output: Number of times 10 appears: 0
````

## List Comprehension

List comprehension is a concise way to create lists in Python. It provides a syntactic construct for creating a list based on existing lists, applying an expression to each element, and optionally filtering elements based on a condition.

Here's the basic syntax for list comprehension:

```
[expression for item in iterable if condition]

```

**Breakdown of the Syntax**
- `expression`: The value or transformation applied to each element in the iterable.
- `item`: The variable that takes the value of each element in the iterable.
- `iterable`: A collection of elements you are iterating over (e.g., list, range, string).
- `condition`: (Optional) A boolean expression that filters elements.

**Examples**

````python
#Simple List Comprehension
# Create a list of squares for numbers from 0 to 9.
squares = [x**2 for x in range(10)]
print(squares)  # Output: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]
````

````python
#List Comprehension with Condition:
# Create a list of even numbers from 0 to 9.
evens = [x for x in range(10) if x % 2 == 0]
print(evens)  # Output: [0, 2, 4, 6, 8]
````

````python
#Nested List comprehension
# Create a list of tuples (x, y) for x in [1, 2, 3] and y in [4, 5, 6].
pairs = [(x, y) for x in [1, 2, 3] for y in [4, 5, 6]]
print(pairs)  # Output: [(1, 4), (1, 5), (1, 6), (2, 4), (2, 5), (2, 6), (3, 4), (3, 5), (3, 6)]
````

````python
#List Comprehension with Function:
#Apply a function to each element in a list.
def double(x):
    return x * 2

doubled = [double(x) for x in range(5)]
print(doubled)  # Output: [0, 2, 4, 6, 8]
````

## Benefits of List Comprehension

- Conciseness: It allows writing more compact and readable code.
- Performance: List comprehensions can be more efficient than using traditional loops.

````python
#Using loop
squares = []
for x in range(10):
    squares.append(x**2)
````

````python
#Using List Comprehension
squares = [x**2 for x in range(10)]
````
