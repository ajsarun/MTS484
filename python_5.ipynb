#Tuple

In Python, a tuple is an **immutable** sequence type, meaning that once it is created, its elements cannot be changed, added, or removed. Tuples are often used to store collections of heterogeneous data and are defined using parentheses `()`.

## **Key Characteristics of Tuples:**
- Ordered: Tuples maintain the order of elements, meaning that the position of each element is fixed.
- Immutable: Once a tuple is created, you cannot modify its elements (i.e., you cannot add, remove, or change elements).
- Heterogeneous: Tuples can store elements of different data types (e.g., integers, strings, lists, etc.).
- Indexable: Elements in a tuple can be accessed using indices, starting from 0 for the first element.

## Creating Tuples

Tuples can be created by placing comma-separated values **(inside parentheses)**:


````python
# Creating a tuple
my_tuple = (1, 'hello', 3.14)

# Single element tuple (note the trailing comma)
single_element_tuple = (42,)

# Without parentheses
another_tuple = 1, 2, 3
print(my_tuple)
print(single_element_tuple)
print(another_tuple)
not_a_tupple = (42)
print(type(not_a_tupple)) #it is an int
print(type(my_tuple))
print(type(single_element_tuple))
print(type(another_tuple))
````

## Create tuples using tuple's constructor

We can create a tuple using the `tuple ()` constructor by passing an iterable object as its argument.

````python
my_tuple = tuple([1, 'hello', 3.14])  #list is an argument
print(my_tuple)
my_tuple1 = tuple('hello')  #string is an argument
print(my_tuple1)
my_tuple2 = tuple((1, 2, 3))  #tuple is an argument
print(my_tuple2)
````

## **Accessing Tuple Elements:**

You can access tuple elements using indexing and slicing:

````python
# Accessing elements
my_tuple = (1, 'hello', 3.14)
print(my_tuple[0])  # Output: 1
print(my_tuple[1:3])  # Output: ('hello', 3.14)
#print(my_tuple[3])  # Raises IndexError
````

## Unpacking a tuple

Assign individual elements to variables:


````python
id, name, age = (1, 'Alice', 25)
print(id)
print(name)
print(age)
````

## Iterating a tuple

Use a loop to go through each item:


````python
my_tuple = (1, 'hello', 3.14)
for item in my_tuple:
    print(item)
````

## Tuple of Tuple

````python
matrix = ((1, 2), (3, 4))
print(matrix)  # Output: [[1, 2], [3, 4]]
print('_' * 20)

#access each element
for row in matrix:
    for item in row:
        print(item)
print('_' * 20)


#access each element using index
for i in range(len(matrix)):
    for j in range(len(matrix[i])):
      print(matrix[i][j], end = " ")
    print()
print('_' * 20)
for row, col in matrix:
    print(row, col)
````

You can see that a Tuple is similar to a List, except a Tuple is immutable.

````python
num_tuple = (1, 2, 3)
num_tuple[0] = 4  # Raises a TypeError
````

*Note:*

If you need to change the value of a Tuple, you can do so as shown in the following code by creating a List first, changing the value, and converting the List back to a Tuple. **However, you must have a very good reason to do so.**

````python
num_tuple = (1, 2, 3)
num_list = list(num_tuple)
num_list[0] = 4
num_tuple = tuple(num_list)
````

## Tuple of Lists

A list is a mutable object. If a tuple contains lists, you can change the values of the lists while the tuple remains the same.

````python
list1 = [1,2,3]
list2 = [4,5,6]
my_tuple = (list1, list2)
print(my_tuple)
#change the value of Lists in a Tuple
list1[0] = 5
list2[2] = 8
print(my_tuple) #The tuple value does not change, but the things that change are the lists' values.
````

## Built-in Python's functions that always applied to a tuple

````python
numbers = (1, 2, 3, 4, 5)
print(len(numbers))
print(sum(numbers))
print(max(numbers))
print(min(numbers))
````

## **Tuple Methods:**

Since tuples are immutable, they have fewer methods compared to lists. Common tuple methods include:

- count(value): Returns the number of times the specified value appears in the tuple.
- index(value): Returns the index of the first occurrence of the specified value.

````python
# Example tuple
my_tuple = (10, 20, 30, 20, 40, 50)

# Find the index of the first occurrence of 20
index_20 = my_tuple.index(20)
print("Index of 20:", index_20)  # Output: Index of 20: 1

# Find the index of 20, starting from index 2
index_20_from_2 = my_tuple.index(20, 2)
print("Index of 20 starting from index 2:", index_20_from_2)  # Output: Index of 20 starting from index 2: 3

# Attempting to find an element not in the tuple
try:
    index_100 = my_tuple.index(100)
except ValueError as e:
    print("Error:", e)  # Output: tuple.index(x): x not in tuple
````

````python
# Example tuple
my_tuple = (1, 2, 3, 2, 4, 2, 5)

# Count the number of times 2 appears in the tuple
count_2 = my_tuple.count(2)
print("Number of times 2 appears:", count_2)  # Output: Number of times 2 appears: 3

# Count the number of times 5 appears in the tuple
count_5 = my_tuple.count(5)
print("Number of times 5 appears:", count_5)  # Output: Number of times 5 appears: 1

# Count the number of times 10 appears in the tuple (which is not present)
count_10 = my_tuple.count(10)
print("Number of times 10 appears:", count_10)  # Output: Number of times 10 appears: 0
````

You can see that `index()` and `count()` of Tuple and List are similar.

## Common Use Cases of Tuple

- Representing Fixed Data: Tuples are great for storing data that shouldn't change, like coordinates, database records, or configuration settings.
- Returning Multiple Values from Functions: Functions can conveniently return multiple values as a tuple.
- Data Integrity: The immutability of tuples helps ensure that data remains consistent throughout your program.

## Why Use Tuples Over Lists

While lists and tuples seem similar, the choice between them depends on your needs:

- If you need to modify the collection, use a list.
- If the data should remain constant and you want to emphasize that it shouldn't be changed, use a tuple. Additionally, tuples can be slightly faster than lists in some operations.
